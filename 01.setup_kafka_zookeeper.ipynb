```
%sh
wget https://downloads.apache.org/kafka/3.8.0/kafka_2.12-3.8.0.tgz -P /dbfs/projeto/
```

```
%sh
/dbfs/projeto/kafka_2.12-3.8.0/bin/zookeeper-server-start.sh /dbfs/projeto/kafka_2.12-3.8.0/config/zookeeper.properties
```
